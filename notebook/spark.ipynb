```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Test").getOrCreate()
```

```python
df = spark.read.csv("Housing.csv", header=True, inferSchema=True)
spark.read
```

```python
df.show()
```

```python
from pyspark.ml.feature import StringIndexer
```

```python

indexer = StringIndexer(inputCols=['guestroom','mainroad','basement','hotwaterheating','airconditioning'], outputCols=['guestroom_ind','mainroad_ind','basement_ind','hotwaterheating_ind'])
df_indexed = indexer.fit(df).transform(df)
```
